Langkah membuat model machine learning :

Hal pertama yang perlu dilakukan yaitu **Import Data** dengan cara mengambil data dari database lain kedalam google collaboratory. Setelah data sudah diimport ke google collaboratory, perlu dilakukan **Data Imputation** untuk mengganti data yang hilang dengan nilai yang disubstitusikan. Selain itu juga perlu dilakukan **Feature Engineering & Selection** untuk menghapus features yang berlebihan dan tidak relevan dari dataset yang sebenarnya. Sebelum melakukan modeling, perlu **Train-test splitting** untuk mengevaluasi performa model machine learning. **Oversampling** juga perlu dilakukan untuk mencegah overfitting model pada fitur target. Setelah semuanya sudah dilakukan, baru **Modeling** machine learning dilakukan. Selanjutnya perlu adanya **Cross-validation** untuk evaluasi kinerja prediktif dari model. Kemudian **Model selection** untuk memilih model machine learning dari serangkaian model yang cocok dengan suatu permasalahan
